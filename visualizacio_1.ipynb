4. ¿Existen relaciones visibles entre la tasa de natalidad y las siguientes variables: tasa de mortalidad infantil, esperanza de vida, densidad de población y población urbana? Usa un subplot. Describe cualquier patrón que encuentres.

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize=(20,10))

sns.regplot(x = "fertility_rate", 
            y = "infant_mortality",
            data = df, 
            color = "turquoise", 
            ax = axes[0,0])

sns.regplot(x = "fertility_rate", 
            y = "life_expectancy",
            data = df, 
            color = "turquoise", 
            ax = axes[0,1])

sns.regplot(x = "fertility_rate", 
            y = "density",
            data = df, 
            color = "turquoise", 
            ax = axes[1,0])

sns.regplot(x = "fertility_rate", 
            y = "urban_population",
            data = df, 
            color = "turquoise", 
            ax = axes[1,1]);
```
