### Paradoja de Simpson

Fenómeno en el que una tendencia que aparece en varios grupos de datos desaparece o incluso se invierte cuando los grupos se combinan. Es un ejemplo clásico de cómo los datos pueden ser manipulados o malinterpretados dependiendo de cómo se agrupan o segmentan. En otras palabras, lo que parece ser una relación consistente en subconjuntos de los datos puede desaparecer o volverse opuesto cuando se analizan los datos globalmente.

```python
sns.scatterplot(x='gdp', y='unemployment_rate', hue ='continent',data=df)
```

```python
sns.regplot(x='gdp', y='unemployment_rate',data=df)
```

¿Cómo se relacionan la tasa de mortalidad infantil y la esperanza de vida?

```python
df.columns
```

```python
sns.scatterplot(x='infant_mortality', y='life_expectancy', data=df)
```

```python
sns.regplot(x='infant_mortality', y='life_expectancy', data=df,  scatter_kws={'color': 'green'}, line_kws={'color': 'red'})
```

En este caso observamos una clara relación entre las variables, a mayor esperanza de vida, menor tasa de mortalidad infantil.
