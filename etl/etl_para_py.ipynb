```python
clientes.columns = clientes.columns.str.title()
clientes.columns
```

```python
productos.head()
```

```python
productos = productos.rename(columns={'ID': 'Id_Producto'})
productos.head(2)
```

```python
productos = productos.rename(columns={'Precio': 'Precio_€'})
productos.head(2)
```

```python
productos["Descripción"] = productos["Descripción"].str.replace(";", ",")
productos.head(2)
```

```python
ventas.head(2)
```

```python
ventas = ventas.rename(columns={'ID_Cliente': 'Id_Cliente'})
ventas.head(2)
```

```python
ventas = ventas.rename(columns={'Total': 'Total_€'})
ventas.head(2)
```

```python
ventas["Fecha_Venta"] = pd.to_datetime(ventas["Fecha_Venta"])
ventas["Fecha_Venta"].dtype
```

```python
# Unir Ventas con Productos por IDproducto
# Unir ventas con clientes por IDCliente
# Primero tenemos que cambiar los nombres de las columnas
```

```python
union= pd.merge(clientes,ventas, on = 'Id_Cliente',how ='inner')
union.head()
```

```python
apertura_exploracion(union)
```

```python
df = pd.merge(union,productos,left_on ='ID_Producto', right_on = 'ID_Producto',how ='left')
df.head()
```
